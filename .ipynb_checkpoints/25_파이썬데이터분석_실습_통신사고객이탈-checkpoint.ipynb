# * Telco Customer Churn Dataset  컬럼

---


1. CustomerID: Customer ID unique for each customer
2. gender: Whether the customer is a male or a female
3. SeniorCitizen: Whether the customer is a senior citizen or not (1, 0) : 고령자 여부
4. Partner: Whether the customer has a partner or not (Yes, No)
5. Dependents: Whether the customer has dependents or not (Yes, No) : 부양가족 여부
6. Tenure: Number of months the customer has stayed with the company : 서비스 사용 개월수
7. PhoneService: Whether the customer has a phone service or not (Yes, No)
8. MultipleLines: Whether the customer has multiple lines or not (Yes, No, No phone service)
9. InternetService: Customer’s internet service provider (DSL, Fiber optic, No)
10. OnlineSecurity: Whether the customer has online security or not (Yes, No, No internet service)
11. OnlineBackup: Whether the customer has an online backup or not (Yes, No, No internet service)
12. DeviceProtection: Whether the customer has device protection or not (Yes, No, No internet service)
13. TechSupport: Whether the customer has tech support or not (Yes, No, No internet service)
14. StreamingTV: Whether the customer has streaming TV or not (Yes, No, No internet service)
15. StreamingMovies: Whether the customer has streaming movies or not (Yes, No, No internet service)
16. Contract: The contract term of the customer (Month-to-month, One year, Two years)
17. PaperlessBilling: Whether the customer has paperless billing or not (Yes, No)
18. PaymentMethod: The customer’s payment method (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic))
19. MonthlyCharges: The amount charged to the customer monthly
20. TotalCharges: The total amount charged to the customer
21. Churn: Whether the customer churned or not (Yes or No) : 이탈 여부

#### 고객 세그먼트별 이탈률 비교



```python
if df['Churn'].dtype != 'object':
    df['Churn'] = df['Churn'].map({1: 'Yes', 0: 'No'})

def churn_rate(df, column):
    return df.groupby(column)['Churn'].apply(lambda x: (x == 'Yes').mean()).sort_values(ascending=False)

segments = ['Contract', 'InternetService', 'gender', 'SeniorCitizen', 'PaymentMethod']
churn_rates = {}

for segment in segments:
    if segment in df.columns:
        churn_rates[segment] = churn_rate(df, segment)
    else:
        print(f"Warning: '{segment}' column not found in the dataframe")
```

```python
if churn_rates:
    plt.figure(figsize=(15, 20))

    for i, (segment, rates) in enumerate(churn_rates.items(), 1):
        plt.subplot(3, 2, i)
        rates.plot(kind='bar')
        plt.title(f'Churn Rate by {segment}')
        plt.xlabel(segment)
        plt.ylabel('Churn Rate')
        plt.xticks(rotation=45, ha='right')
        
        # 막대 위에 비율 표시
        for j, v in enumerate(rates):
            plt.text(j, v, f'{v:.2%}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```

- female, male 은  비슷하다.
- 달마다 결제하는 고객의 이탈률이 높음
- SeniorCitizen 의 비율이 높음.

장기 고객자 세그먼트별 비교 

```python
# 3년 이상 사용한 고객 
long_term_customers = df[df['tenure'] >= 36]


# 세그먼트별 장기 고객 비율 계산 함수
def long_term_customer_rate(df, column):
    return df.groupby(column)['tenure'].apply(lambda x: (x >= 36).mean()).sort_values(ascending=False)

# 세그먼트 정의
segments = ['Contract', 'InternetService', 'gender', 'SeniorCitizen', 'PaymentMethod']

# 세그먼트별 장기 고객 비율 계산
long_term_rates = {}
for segment in segments:
    if segment in df.columns:
        long_term_rates[segment] = long_term_customer_rate(df, segment)
    else:
        print(f"Warning: '{segment}' column not found in the dataframe")

# 결과 시각화
plt.figure(figsize=(15, 20))

for i, (segment, rates) in enumerate(long_term_rates.items(), 1):
    plt.subplot(3, 2, i)
    rates.plot(kind='bar')
    plt.title(f'Long-term Customer Rate by {segment}')
    plt.xlabel(segment)
    plt.ylabel('Long-term Customer Rate')
    plt.xticks(rotation=45, ha='right')
    
    # 막대 위에 비율 표시
    for j, v in enumerate(rates):
        plt.text(j, v, f'{v:.2%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

이탈하지 않은 장기고객자 

```python
# Define long-term customers who have not churned (tenure >= 36 months and Churn is 'No')
long_term_non_churned_customers = df[(df['tenure'] >= 36) & (df['Churn'] == 'No')]

# Function to calculate the proportion of long-term non-churned customers by segment
def segment_distribution(df, column):
    return df[column].value_counts(normalize=True).sort_values(ascending=False)

# Define segments to analyze
segments = ['Contract', 'InternetService', 'gender', 'SeniorCitizen', 'PaymentMethod']

# Calculate segment distribution for long-term non-churned customers
segment_distributions = {}
for segment in segments:
    if segment in df.columns:
        segment_distributions[segment] = segment_distribution(long_term_non_churned_customers, segment)
    else:
        print(f"Warning: '{segment}' column not found in the dataframe")

# Visualize the results
plt.figure(figsize=(15, 20))

for i, (segment, distribution) in enumerate(segment_distributions.items(), 1):
    plt.subplot(3, 2, i)
    distribution.plot(kind='bar')
    plt.title(f'Distribution of Long-term Non-churned Customers by {segment}')
    plt.xlabel(segment)
    plt.ylabel('Proportion')
    plt.xticks(rotation=45, ha='right')
    
    # Display proportion on top of bars
    for j, v in enumerate(distribution):
        plt.text(j, v, f'{v:.2%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

고객 이탈자 세그먼트별 비교 

```python

# 이탈 고객 정의 (Churn 열이 'Yes'인 경우)
churned_customers = df[df['Churn'] == 'Yes']

# 세그먼트별 이탈 고객 비율 계산 함수
def churn_rate_by_segment(df, column):
    return df.groupby(column)['Churn'].apply(lambda x: (x == 'Yes').mean()).sort_values(ascending=False)

# 세그먼트 정의
segments = ['Contract', 'InternetService', 'gender', 'SeniorCitizen', 'PaymentMethod']

# 세그먼트별 이탈 고객 비율 계산
churn_rates = {}
for segment in segments:
    if segment in df.columns:
        churn_rates[segment] = churn_rate_by_segment(df, segment)
    else:
        print(f"Warning: '{segment}' column not found in the dataframe")

# 결과 시각화
plt.figure(figsize=(15, 20))

for i, (segment, rates) in enumerate(churn_rates.items(), 1):
    plt.subplot(3, 2, i)
    rates.plot(kind='bar')
    plt.title(f'Churn Rate by {segment}')
    plt.xlabel(segment)
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45, ha='right')
    
    # 막대 위에 비율 표시
    for j, v in enumerate(rates):
        plt.text(j, v, f'{v:.2%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

데이터에 결측치 있으면 모델링시 알지 못하는 에러 발생할수 있으므로 반드시 결측치를 제거나 변경해야 한다.<br>
